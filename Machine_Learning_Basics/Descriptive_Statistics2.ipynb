```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
%matplotlib inline
```

```python
cars=pd.read_csv('mtcars.csv')
cars
```

```python
car1=cars.drop('Unnamed: 0',axis=1)
car1
```

###  Measure of Center Tendecy       
 Mean,
 Meadian,  
 Mode

                        

```python
car1.mean()
```

```python
car1.median()
```

```python
car1.mode()
```

```python
#Range
max(car1['mpg'])-min(car1['mpg'])
```

```python
#Quantile values
q_value=[car1['mpg'].quantile(0),
        car1['mpg'].quantile(0.25),
        car1['mpg'].quantile(0.50),
         car1['mpg'].quantile(0.75),
        car1['mpg'].quantile(1)]
```

```python
q_value
```

### Describe()

#### This function return inner quantile values and describe data

```python
car1['mpg'].describe()
```

### Interquartile Range

```python
car1['mpg'].quantile(0.75)-car1['mpg'].quantile(0.25)
```

```python
car1.boxplot(column='mpg',return_type='axes',
            figsize=(8,8))

py.text(x=0.74,y=22.25,s='3rd quartile')
py.text(x=0.8,y=18.75,s='median')
py.text(x=0.75,y=15.5,s='1st quartile')
py.text(x=0.9,y=10,s='min')
py.text(x=0.9,y=33.5,s='max')
py.text(x=0.7,y=19.5,s='IQR',rotation=90,size=25);
```
